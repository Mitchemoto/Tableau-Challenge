```python
#import dependencies
import pandas as pd
import numpy as np
import glob
import os
```

```python
#bring in file path
path = r'Resources'
#use glob to read through all files
all_files=glob.glob(path + "/*.csv")
#create empty list to fill with csv data
li =[]
#create for loop to reach through each file and append to li
for filename in all_files:
    bikeDataDf=pd.read_csv(filename, index_col=None, header=0)
    li.append(bikeDataDf)

frame = pd.concat(li, axis=0, ignore_index=True)
frame.head()
```

```python
#concat start station name and end station name to get unique paths
cols=['start station name','end station name']

frame['path_concat']=frame[cols].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)

frame['path_concat'].head()
```

```python
frame
```

```python
frame['path_concat']=frame['path_concat'].str.replace(' ','-')
frame['path_concat']
```

```python
frame
```

```python
#convert frame to csv with path_concat column to use as relationship for origin-destination df
frame.to_csv('Data\CombinedData.csv')
```

```python
#single out the columns needed
startStationDf=frame[['start station id', 'start station name','start station latitude','start station longitude','path_concat']]

startStationDf.head()
```

```python
#insert origin-destination marker
startStationDf.insert(5,"Origin-Destination","Origin")
originDf=startStationDf
originDf.head()
```

```python
#clean header data so can merge with destination
cleanOriginDf=originDf.rename(columns={
                "start station id":"station id",
                "start station name":"station name",
                "start station latitude":"station latitude",
                "start station longitude":"station longitude",
                    })
cleanOriginDf.head()
```

```python
#single out the columns needed
endStationDf=frame[['end station id','end station name','end station latitude','end station longitude','path_concat']]
endStationDf.head()
```

```python
#insert destination marker
endStationDf.insert(5,"Origin-Destination","Destination")
destinationDf=endStationDf
destinationDf.head()
```

```python
#clean header data so can merge with origin
cleanDestinationDf=destinationDf.rename(columns={
                "end station id":"station id",
                "end station name":"station name",
                "end station latitude":"station latitude",
                "end station longitude":"station longitude",
                    })
cleanDestinationDf.head()
```

```python
#drop duplicates in path_concat to get individual 
ddStartDf=cleanOriginDf.drop_duplicates(subset=['path_concat'],keep='last',ignore_index=False)
ddStartDf
```

```python
# #drop duplicates in path_concat to get individual 
ddEndDf=cleanDestinationDf.drop_duplicates(subset=['path_concat'],keep='last',ignore_index=False)

ddEndDf
```

```python
frames1=[ddStartDf, ddEndDf]
indexDf=pd.concat(frames1)
indexDf
```

```python
#verify that the index match before resetting so they are matched between the 
sortedODDf=indexDf.sort_values(['path_concat','Origin-Destination'], ascending=[True,False])
sortedODDf
```

```python
# reset index
sortedResetDf=sortedODDf.reset_index()
sortedResetDf
```

```python
#need to add in the dash between string values in station name for Tableau formula to work
sortedResetDf['station name']=sortedResetDf['station name'].str.replace(' ','-')
```

```python
#drop 0 value lat/long as error occured
droperrordf=sortedResetDf[sortedResetDf['station latitude']!=0]
```
